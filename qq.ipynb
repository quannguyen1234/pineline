```python
import random
def tinh():
    von=100000
    solan=1
    origin=100000
    goc=1000

    while solan!=10:
        luachon_may=random.choice([1,2])
        luachon_nguoi=random.choice([1,2])
        if luachon_may==luachon_nguoi:
            von+=goc
            goc=1000
        else:
            von-=goc
            goc=goc*2
        solan+=1
        
        if von <=0:
            return False   
    print(von,origin) 
    if von<origin:
        return False
    return True

count=0
for i in range(10000):
    if tinh():
        count+=1
    # break
print(count)
```
